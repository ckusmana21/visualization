*   The variables I chose are State (I wanted to see if and how the distribution of treatment programs differ across states); NPI (this seems to be the unique indentifier, so I can use this to check for uniqueness and NAs); Medicare ID Effective Date (I wanted to see if there are any time trends with how long the treatment providers have been in the system).




```python
opioid = pd.read_csv('/content/OPIOID_TREATMENT_PROGRAM_PROVIDERS_09302024.csv')
#opioid.head()

columns_of_interest = ['NPI', 'STATE', 'MEDICARE ID EFFECTIVE DATE']
df = opioid[columns_of_interest]

filtered_columns = ['NPI', 'STATE', 'MEDICARE ID EFFECTIVE DATE']
filtered_opioid = opioid[columns_of_interest]

# 1. Cleaning NPI column; checking any missing values and uniqueness (since it's a unique identifier)
filtered_opioid['NPI'] = filtered_opioid['NPI'].astype(str).str.strip()
filtered_opioid.dropna(subset=['NPI'], inplace=True)
# NOTE: I used ChatGPT to display the output, resulting in the script below and replicated across the other 2 columns:
print(f"Number of unique NPIs: {filtered_opioid['NPI'].nunique()}")

# 2. Cleaning STATE column; checking for consistency in abbreviations
filtered_opioid['STATE'] = filtered_opioid['STATE'].str.upper().str.strip()
print(f"Unique states: {filtered_opioid['STATE'].unique()}")

# 3. Cleaning MEDICARE ID EFFECTIVE DATE; converting the date column to datetime format
# NOTE: I used ChatGPT's help with the script below:
filtered_opioid['MEDICARE ID EFFECTIVE DATE'] = pd.to_datetime(filtered_opioid['MEDICARE ID EFFECTIVE DATE'], errors='coerce')

# 4. Cleaning NAs, dropping rows where the date couldn't be converted
filtered_opioid.dropna(subset=['MEDICARE ID EFFECTIVE DATE'], inplace=True)
print(filtered_opioid['MEDICARE ID EFFECTIVE DATE'].describe())
```

4.4

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar graph for STATE (to see the distribution of teratment program across states)
plt.figure(figsize=(12, 8))
filtered_opioid['STATE'].value_counts().plot(kind='bar')
plt.title('Bar Graph of States with Opioid Treatment Providers')
plt.xlabel('STATE')
plt.ylabel('Count')
plt.show()

# Checking NPI (to see how many unique observations)
print(f"Number of unique NPIs: {filtered_opioid['NPI'].nunique()}")

# Boxplot for MEDICARE ID EFFECTIVE DATE (to see any outliers)
filtered_opioid['MEDICARE ID EFFECTIVE DATE'] = pd.to_datetime(filtered_opioid['MEDICARE ID EFFECTIVE DATE'], errors='coerce')

filtered_opioid['Medicare Year'] = filtered_opioid['MEDICARE ID EFFECTIVE DATE'].dt.year

# Create a boxplot for 'Medicare Year'
plt.figure(figsize=(10, 6))
sns.boxplot(y=filtered_opioid['Medicare Year'])
plt.title('Boxplot of Medicare ID Effective Year')
plt.ylabel('Medicare Year')
plt.show()

```
